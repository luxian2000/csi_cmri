```python
import pennylane as qml
import torch
import torch.nn as nn
import numpy as np
import time
import os
```

```python
torch.manual_seed(42)
np.random.seed(42)
```

```python
data_30 = np.load('../DataSpace/csi_cmri/CSI_channel_30km.npy')  # shape=(80000, 2560)
```

```python
dev = qml.device("default.qubit", wires=2)
```

```python
@qml.qnode(dev, interface = "torch")
def cir(x, w):
    y = torch.matmul(x, w)
    qml.RX(y, wires=0)
    qml.CNOT(wires=[1,0])
    qml.RY(y, wires=1)
    return qml.expval(qml.PauliZ(wires=0))
```

```python
x = torch.tensor(np.array([1.0, 2.0]), requires_grad=True)
w = torch.tensor(np.array([2.0, 3.0]), requires_grad=True)

result = cir(x, w)
result.backward()
x.grad
```
